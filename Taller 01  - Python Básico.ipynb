```python
## Haz tu codigo aquí:
detalle_venta2 = {

        "producto": "camisa",
        "categoria": "Ropa",
        "precio": (precio2 := 30000),
        "cantidad": (cantidad2 := 3),
        "total": (total_venta2 := precio2 * cantidad2)

}

if detalle_venta2 not in registro_ventas: 
    registro_ventas.append(detalle_venta2)

detalle_venta3 = {

        "producto": "pizza",
        "categoria": "Alimentos",
        "precio": (precio3 := 10000),
        "cantidad": (cantidad3 := 5),
        "total": (total_venta3 := precio3 * cantidad3)

}

if detalle_venta3 not in registro_ventas: 
    registro_ventas.append(detalle_venta3)


for venta in registro_ventas: 
    total_original = venta["precio"]
    producto_actual = venta["producto"]

    if total_original > 100000:
        total_final = total_original - (total_original * 0.1)
        print("Producto:", producto_actual)
        print("El precio total original es:", total_original)
        print("precio final con descuento es", total_final)

    else: 
        print("Producto:", producto_actual)
        print("el precio final es", total_original)
```

```python
## Haz tu codigo aquí:

def calcular_ingreso (precio, cantidad):
    try: 
        total_ingreso = int(precio) * int(cantidad)
        return total_ingreso
    
    except ValueError: 
        print("Error: Los valores deben ser numéricos")
        return 0
    
ingreso1 = calcular_ingreso(input("Ingrese el precio"), input("Ingrese la cantidad"))
print("El ingreso total es de", ingreso1)
    


    





```
